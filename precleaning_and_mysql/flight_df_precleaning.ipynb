```python
import pandas as pd
import time
from datetime import datetime
import pymysql

csv_path = 'flight_df2.csv'
```

```python
# csv 파일 불러오기
flight_df = pd.read_csv(csv_path)
```

```python
flight_df
```

```python
# 컬럼 앞에 모두 띄어 쓰기가 되어 있어서 컬럼명 재설정. 
flight_df.columns = ['Date', 
                     'Day', 
                     'Airline', 
                     'Departure time', 
                     'Departure airport',
                     'Arrival time', 
                     'Arrival airport', 
                     'Flight info', 
                     'Flight time',
                     'Price'
                    ]
```

```python
# flight_df 유니크 값 모두 보기 
print(flight_df['Date'].unique())
print(flight_df['Day'].unique())
print(flight_df['Airline'].unique())
print(flight_df['Departure time'].unique())
print(flight_df['Departure airport'].unique())
print(flight_df['Arrival time'].unique())
print(flight_df['Arrival airport'].unique())
print(flight_df['Flight info'].unique())
print(flight_df['Flight time'].unique())
print(flight_df['Price'].unique())
```

```python
type(flight_df['Date'][0])
```

```python
# 6.10 형식에서 20230610 형식으로 int타입으로 변경
flight_df['Date'] = flight_df['Date'] * 100 + 20230000
```

```python
flight_df['Date'] = flight_df['Date'].astype('int')
```

```python
flight_df['Date'].unique()
```

```python
# 직항만 있는 Flight info 열 제거 
flight_df.drop(columns='Flight info', axis=1, inplace=True)
```

```python
# Airline ANA = ANA항공 이므로 ANA를 ANA항공으로 변경 
flight_df.loc[flight_df['Airline'] == 'ANA', 'Airline'] = 'ANA항공'
```

```python
# Price 열 int 타입으로 변경 
for i, price in enumerate(flight_df['Price']):
    flight_df.loc[i, 'Price'] = str(price).replace(',', '')
    
flight_df['Price'] = flight_df['Price'].astype('int')
```

```python
# 20230601 -> 2023-06-01 형식으로 변환
for i, date in enumerate(flight_df['Date']):
    flight_df.loc[i, 'Date'] = str(date)[:4] + '-' + str(date)[4:6] + '-' + str(date)[6:]
```

```python
# datetime 타입으로 변경 
flight_df['Date'] = pd.to_datetime(flight_df['Date'])
```

```python
# 시간:분: 초 단위로 변경 
for i, time in enumerate(flight_df['Departure time']):
    flight_df.loc[i, 'Departure time'] = time + ':00'
```

```python
# 시간:분: 초 단위로 변경 
for i, time in enumerate(flight_df['Arrival time']):
    flight_df.loc[i, 'Arrival time'] = time + ':00'
```

```python
# index 열 (id 열) 추가 
flight_df = flight_df.reset_index()
```

```python
# 비행 시간 hh:mm:ss 형태로 변경

for i, time in enumerate(flight_df['Flight time']):
    flight_df.loc[i, 'Flight time'] = flight_df['Flight time'][i][:2] + ':' + flight_df['Flight time'][i][5:7] +':00'
```

```python
flight_df
```

```python
flight_df.to_csv('precleaning_flight_df.csv', index=None)
```
