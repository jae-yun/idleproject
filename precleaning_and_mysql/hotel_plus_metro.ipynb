```python
date_list = [
             '601', '602', '603', '604', '605', '606', '607', '608', '609', '610', 
             '611', '612', '613', '614', '615', '616', '617', '618', '619', '620', 
             '621', '622', '623', '624', '625', '626', '627', '628', '629', '630',
             '701', '702', '703', '704', '705', '706', '707', '708', '709', '710', 
             '711', '712', '713', '714', '715', '716', '717', '718', '719', '720',
             '721', '722', '723', '724', '725', '726', '727', '728', '729', '730',
            ]
```

```python
df.dropna(inplace=True)
```

```python
df = df.reset_index(drop=True)
```

```python
empty_price_idx = []

for i in range(len(df)):
    num_not_zero = 0
    sum_price = 0
   
    for date in date_list:
        if df[date][i] != 0:
            num_not_zero += 1
            sum_price += df[date][i]
            
    if num_not_zero == 0:
        empty_price_idx.append(i)
```

```python
empty_price_idx
```

```python
df.drop(empty_price_idx, axis=0, inplace=True)
```

```python
df = df.reset_index(drop=True)
```

```python
empty_price_idx = []

for i in range(len(df)):
    num_not_zero = 0
    sum_price = 0
   
    for date in date_list:
        if df[date][i] != 0:
            num_not_zero += 1
            sum_price += df[date][i]
            
    df.loc[i, 'mean_price'] = int(sum_price / num_not_zero)
```
