```python
def merge_address_score(df):
    for idx in df[df['Hotel Address_x'].isna()].index:
        df['Hotel Address_x'][idx] = df['Hotel Address_y'][idx]
        
    for idx in df[df['Hotel Score_x'].isna()].index:
        df['Hotel Score_x'][idx] = df['Hotel Score_y'][idx]
        
    return df

df = merge_address_score(df)
```
